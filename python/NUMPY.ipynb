## **CREATING THE FILTER ARRAY**

Create a filter array that will return only values higher than 42:

```python
import numpy as np
arr=np.array([41,42,43,44])

# create an empty list
filter_arr = []

# go through each element in arr
for element in arr:

  # if the element is higher than 42,set the value to True, otherwise False:
  if element>42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

create a filter array that will return only even elements from the original array:

```python
import numpy as np
arr=np.array([1,2,3,4,5,6,7])

filter_arr = []
for element in arr:
  if element % 2 == 0:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

## **CREATING FILTER DIRECTLY FROM ARRAY**

Create a filter array that will return only values higher than 42:

```python
import numpy as np
arr=np.array([41,42,43,44])
filter_arr = arr >42
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

create a filter array that will return only even elements from the original array:

```python
import numpy as np
arr = np.array([1,2,3,4,5,6,7])
filter_arr = arr % 2 == 0
newarr = arr[filter_arr]
print(filter_arr)
print(newarr)
```

program 2

```python
import numpy as np
import scipy.stats as stats

data = [12,15,14,10,18,20,22,25,14,17]

#calculate range
range_value = max(data)-min(data)
print(f"Range : {range_value}")
```

```python
#calculate variance
variance_value = np.var(data,ddof=0) # ddof=0-population variance

sample_variance_value = np.var(data,ddof=1) # ddof=1-sample variance

print(f"Population Variance : {variance_value}")
print(f"Sample Variance : {sample_variance_value}")
```

```python
#calculate standard deviation
std_dev_value = np.std(data,ddof=0) # population standard deviation
sample_std_dev_value = np.std(data,ddof=1) # sample standard deviation

print(f"Population Standard Deviation : {std_dev_value}")
print(f"Sample Standard Deviation : {sample_std_dev_value}")
```

```python
# calculate IQR value
Q1 = np.percentile(data,25) #first quartile(25th percentile)
Q3 = np.percentile(data,75) #third quartile(75th percentile)

iqr_value = Q3 - Q1

print(f"Interquartile Range (IQR) : {iqr_value}")
```

```python
from numpy import random
x=random.chisquare(df=2 , size=(2,3))
print(x)
```

# **VISUALIZATION OF CHI SQUARE DISTRIBUTION**

# **NUMPY UFUNCS**(UNIVERSAL FUNCTIONS)

# **add the elements of two lists**

without ufunc we can use python built-in zip() method

```python
x = [1,2,3,4]
y = [4,5,6,7]
z = []
for i,j in zip(x,y):
  z.append(i+j)
print(z)

```

with ufunc we can use add() function:

```python
import numpy as np

x = [1,2,3,4]
y = [4,5,6,7]
z = np.add(x,y)
print(z)
```

# create your own ufunc

frompyfunc() method

create your own ufunc for addition:

```python
import numpy as np
def myadd(x,y):
  return x+y
myadd = np.frompyfunc(myadd,2,1)
print(myadd([1,2,3,4],[5,6,7,8]))
```

# **NUMPY SUMMATIONS**

Add the values in arr1 to the values in arr2:

```python
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.add(arr1,arr2)
print(newarr)
```

sum the values in arr1 and the values in arr2:

```python
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.sum([arr1,arr2]) # sum means add all elements in two arrays
print(newarr)
```

## **summation over an axis**

```python
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.sum([arr1,arr2],axis=1) # here sum of each arrays.
print(newarr)
```

## **cumulative sum**

```python
import numpy as np
arr = np.array([1,2,3])
newarr = np.cumsum(arr)  # cumsum means partial sum.ie., here 1 , 1+2 , 1+2+3
print(newarr)
```

# **PRODUCTS**

```python
import numpy as np
arr = np.array([1,2,3,4])
x = np.prod(arr)  # product means 1*2*3*4 (multiply whole elements in the array)
print(x)
```

find the product of elements of two arrays:

```python
import numpy as np
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])
x = np.prod([arr1,arr2]) # 1*2*3*4*5*6*7*8
print(x)
```

## **product over an axis**

```python
import numpy as np
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])
x = np.prod([arr1,arr2],axis=1) # over axis product of each arrays.
print(x)
```

## **cumulative product**

```python
import numpy as np
arr = np.array([5,6,7,8])
newarr = np.cumprod(arr) # 5 , 5*6 , 5*6*7 , 5*6*7*8
print(newarr)
```

# **NUMPY DIFFERENCES**

Compute discrete difference of the following array:

```python
import numpy as np
arr = np.array([10,15,25,5])
newarr = np.diff(arr) # 15-10 , 25-15 , 5-25
print(newarr)
```

Compute discrete difference of the following array twice:

```python
import numpy as np
arr = np.array([10,15,25,5])
newarr = np.diff(arr,n=2)  # 15-10 , 25-15 , 5-25 = [5,10,-20] again 10-5 , -20-10
print(newarr)
```

# **NUMPY LCM (LOWEST COMMON MULTIPLE)**

Find the lcm of following two elements:

```python
import numpy as np
num1 = 4
num2 = 6
x = np.lcm(num1,num2)
print(x)
```

## **find the lcm in arrays**

reduce() method:

```python
import numpy as np
arr = np.array([3,6,9])
x = np.lcm.reduce(arr) # reduce() method: here first find lcm(3,6)=6 then find lcm(6,9) that is 18.
print(x)
```

find the lcm of all of an array where the array contains all integers from 1 to 10:

```python
import numpy as np
arr = np.arange(1,11)
x = np.lcm.reduce(arr)
print(x)
```

# **NUMPY GCD ( GREATEST COMMON DENOMINATOR)**

```python
import numpy as np
num1 = 6
num2 = 9
x = np.gcd(num1,num2)
print(x)
```

## **finding gcd in arrays**

reduce() method:

```python
import numpy as np
arr = np.array([20,8,32,36,16])
x = np.gcd.reduce(arr) # reduce() method : first  gcd(20,8) then gcd(first answer,32) etc...
print(x)
```
